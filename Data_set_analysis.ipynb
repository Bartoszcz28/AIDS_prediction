```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aids_t = pd.read_csv('data/AIDS_Classification_50000.csv')

# aids_t = pd.read_csv('data/aids_t.csv')
```

```python
# 1. Rozkad zmiennych kategorycznych
categorical_columns = ['hemo', 'homo', 'drugs', 'oprior', 'z30', 'race', 'gender', 'str2', 'symptom', 'treat']
plt.figure(figsize=(20, 10))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(2, 5, i)
    sns.countplot(x=column, data=aids_t)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.savefig('images/rozkad_zmiennych_kategorycznych.png')
plt.show()
```

```python
# 2. Rozkad zmiennych numerycznych
numerical_columns = ['preanti', 'cd40', 'cd420', 'cd80', 'cd820']
# numerical_columns = ['numerical__preanti', 'numerical__cd40', 'numerical__cd420', 'numerical__cd80', 'numerical__cd820']
plt.figure(figsize=(20, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(aids_t[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.savefig('images/rozkad_zmiennych_numerycznych.png')
plt.show()
```

```python
# 3. Macierz korelacji
correlation = aids_t.corr().round(2)
plt.figure(figsize=(14, 7))
sns.heatmap(correlation, annot=True, cmap='BuPu')
plt.title('Correlation Matrix')
plt.savefig('images/macierz_korelacji.png')
plt.show()
```

```python
# 4. Wykresy skrzynkowe (box plots)
plt.figure(figsize=(20, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='gender', y=column, data=aids_t)  # Gender as an example of categorical variable
    plt.title(f'Boxplot of {column} by Gender')
plt.tight_layout()
plt.savefig('images/wykresy_skrzynkowe.png')
plt.show()
```

```python
# 5. Wykresy par (pair plots)
sns.pairplot(aids_t[numerical_columns])
plt.savefig('images/wykresy_par.png')
plt.show()
```

```python
# 6. Wykresy punktowe (scatter plots)
plt.figure(figsize=(20, 10))
for i, column in enumerate(numerical_columns[1:], 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(x=numerical_columns[0], y=column, data=aids_t)
    plt.title(f'Scatter Plot of {numerical_columns[0]} vs {column}')
plt.tight_layout()
plt.savefig('images/wykresy_punktowe.png')
plt.show()
```

```python
# 7. Wykresy gstoci (density plots)
plt.figure(figsize=(20, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.kdeplot(aids_t[column], shade=True)
    plt.title(f'Density Plot of {column}')
plt.tight_layout()
plt.savefig('images/Wykresy_gestosci.png')
plt.show()
```
